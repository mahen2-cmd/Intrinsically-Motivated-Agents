```python
def read_images_from_folder(folder_path):
    image_paths = glob.glob(f"{folder_path}/*.png")
    image_paths.sort(key=lambda x: int(re.findall(r"(\d+)", x)[0]))
    images = [Image.open(path) for path in image_paths]
    return images
```
